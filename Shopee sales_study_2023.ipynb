```python
#3 Show number of listing product based on location. You can extract this from specification” field
```

```python
print(df['specification'].head(2))
```

```python
#extract location in specification column
print(df['specification'].unique()[:20])

#check if if 'Selangor', 'KL', 'Penang' etc appear
print(df['specification'].str.contains('Selangor', na=False).sum())
print(df['specification'].str.contains('Kuala Lumpur', na=False).sum())
```

```python
#extract location using Regex

import re

malaysia_locations= [
    'Kuala Lumpur', 'Selangor', 'Penang', 'Johor', 'Sabah',
    'Sarawak', 'Perak', 'Kedah', 'Kelantan', 'Melaka',
    'Negeri Sembilan', 'Pahang', 'Perlis', 'Terengganu', 'Putrajaya','Labuan'
]


overseas_keywords = [
    'China', 'Taiwan', 'Singapore', 'Thailand', 'Indonesia',
    'Japan', 'Korea', 'Hong Kong', 'USA', 'United States',
    'United Kingdom', 'Australia', 'India'
]


#Build regex pattern
malaysia_pattern = '|'.join(malaysia_locations)
overseas_pattern = '|'.join(overseas_keywords)

def classify_location(text):
    if pd.isna(text):
       return 'Unknown'
    text_title=text.title()
    if re.search(malaysia_pattern, text_title, re.IGNORECASE):
        match = re.search(malaysia_pattern, text_title, re.IGNORECASE)
        return match.group(0).title()
    elif re.search(overseas_pattern, text_title,re.IGNORECASE):
        return 'Oversea'
    else:
        return 'Unknown'

#Extract location
df['extracted_location']=df['specification'].apply(classify_location)

location_count=df['extracted_location'].value_counts().reset_index()
location_count.columns = ['location' , 'product_count']

print(location_count)
```

```python
#Visualization for Location Sale

location_count = df['extracted_location'].value_counts().reset_index()
location_count.columns = ['location', 'product_count']

print(location_count)

# Plot
plt.figure(figsize=(12, 5))
plt.bar(location_count['location'], location_count['product_count'], color='steelblue')
plt.title('Number of Product Listings by Location')
plt.xlabel('Location')
plt.ylabel('Number of Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#4 SPlit category details

print(df['item category detail'].head(20))
```

```python
#split the category details into 3 section

category_split = df['item category detail'].str.split('|',expand = True)
print(category_split.head())
```

```python
#make the index 0= shoppe, 1 = Main Category, 2 = Subcategory 1, 3 = Subcategory 2

df['main_category']=category_split[1].str.strip()
df['subcategory_1']=category_split[2].str.strip()
df['subcategory_2']=category_split[3].str.strip()

print(df[['item category detail','main_category', 'subcategory_1', 'subcategory_2']].head(10))
```

```python
#For the top 3 main categories, show the top 5 subcategory 1 for that main categoryn based on number of products

top3_main = df['main_category'].value_counts().head(3).index.tolist()
print(top3_main)
```

```python
#filter only top  main categories
df_top3=df[df['main_category'].isin(top3_main)]

#count subcategory_1 within each main category
top5_subcat=(df_top3.groupby(['main_category','subcategory_1'])
                    .size()
                    .reset_index(name='product_count')
                    .sort_values(['main_category','product_count'],ascending=[True,False]))

#Keep only top 5 subcategories per main category
top5_subcat=top5_subcat.groupby('main_category').head(5)
print(top5_subcat)
```

```python
#Vizualization main category and product count

main_cat_count = df['main_category'].value_counts().reset_index()
main_cat_count.columns = ['main_category', 'product_count']

print(main_cat_count)

# Plot
plt.figure(figsize=(14, 6))
plt.bar(main_cat_count['main_category'], main_cat_count['product_count'], color='mediumpurple')
plt.title('Number of Product Listings by Main Category')
plt.xlabel('Main Category')
plt.ylabel('Number of Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#5  Show price range for each main category

print(df[['main_category','price_actual']].head(10))
print(df['price_actual'].isnull().sum())
```

```python
#6 Show the revenue for each main category in descending order

print(df[['main_category','price_actual', 'total_sold']])

# convert to numeric, error='coerce; turns any invalid values to NaN
df['price_actual'] = pd.to_numeric(df['price_actual'], errors='coerce')
df['total_sold'] = pd.to_numeric(df['total_sold'], errors='coerce')



# Check dtypes are now numeric
print(df[['price_actual', 'total_sold']].dtypes)

#Fill NaN with 0 to avoid missing values in calculation
df['price_actual'] = df['price_actual'].fillna(0)
df['total_sold'] = df['total_sold'].fillna(0)


#revenue=price_actual x total_sold
df['revenue']=df['price_actual']*df['total_sold']

#sum revenue per main category
revenue_per_category=(df.groupby('main_category')['revenue']
                        .sum()
                        .reset_index()
                        .sort_values('revenue',ascending=False))
print(revenue_per_category)
```

```python
plt.figure(figsize=(14, 6))
bars = plt.barh(revenue_per_category['main_category'], 
                revenue_per_category['revenue'], 
                color='steelblue')

# Add value labels
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2,
             f'RM {width:,.0f}',
             ha='left', va='center', fontsize=8)

plt.title('Total Revenue by Main Category', fontsize=14, fontweight='bold')
plt.xlabel('Revenue (RM)')
plt.ylabel('Main Category')
plt.tight_layout()
plt.show()
```
